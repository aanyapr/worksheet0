```python
import numpy as np
arr1 = np.arange(10, 50)
print("Array 10 to 49:\n", arr1)
arr2 = np.arange(9).reshape(3, 3)
print("\n3x3 Matrix 0 to 8:\n", arr2)
arr3 = np.eye(3)
print("\nIdentity Matrix:\n", arr3)
arr4 = np.random.random(30)
print("\nMean of random array:\n", arr4.mean())
arr5 = np.random.random((10, 10))
print("\nMin:", arr5.min(), " Max:", arr5.max())
arr6 = np.zeros(10)
arr6[4] = 1
print("\nZero array with 5th element = 1:\n", arr6)
arr = np.array([1, 2, 0, 0, 4, 0])
arr7 = arr[::-1]
print("\nReversed array:\n", arr7)
arr8 = np.ones((5, 5))
arr8[1:-1, 1:-1] = 0
print("\n2D array with border 1 and inside 0:\n", arr8)
arr9 = np.zeros((8, 8))
arr9[1::2, ::2] = 1
arr9[::2, 1::2] = 1
print("\n8x8 Checkerboard pattern:\n", arr9)
```

```python
import numpy as np
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])
add_xy = x + y
print("\nAddition of x and y:\n", add_xy)
sub_xy = x - y
print("\nSubtraction of x and y:\n", sub_xy)
mul_x = x * 3
print("\nx multiplied by 3:\n", mul_x)
sq_x = x ** 2
print("\nSquare of each element in x:\n", sq_x)
dot_vw = np.dot(v, w)
dot_xv = np.dot(x, v)
dot_xy = np.dot(x, y)
print("\nDot product (v . w):", dot_vw)
print("Dot product (x . v):\n", dot_xv)
print("Dot product (x . y):\n", dot_xy)
concat_xy_rows = np.vstack((x, y))
print("\nConcatenate x & y along rows:\n", concat_xy_rows)
concat_vw_columns = np.column_stack((v, w))
print("\nConcatenate v & w along columns:\n", concat_vw_columns)
print("\nAttempting to concatenate x and v:")
try:
    invalid_concat = np.concatenate((x, v))
except ValueError as e:
    print("Error:", e)
    print("Reason: x is 2D but v is 1D — dimensions do not match.")
v_reshaped = v.reshape(1, 2)
valid_concat = np.concatenate((x, v_reshaped))
print("\nCorrect concatenation after reshaping v:\n", valid_concat)
```

```python
import numpy as np
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

print("Matrix Proofs")
A_inv = np.linalg.inv(A)
A_times_A_inv = np.dot(A, A_inv)
I = np.eye(2)

print("\n1. Prove A * A^-1 = I")
print("Matrix A:\n", A)
print("Inverse of A (A^-1):\n", A_inv)
print("A * A^-1:\n", A_times_A_inv)
print("Identity Matrix I:\n", I)
is_identity = np.allclose(A_times_A_inv, I)
print(f"Is A * A^-1 numerically close to I? {is_identity}")
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n2. Prove AB != BA")
print("A * B (AB):\n", AB)
print("B * A (BA):\n", BA)
is_not_equal = not np.array_equal(AB, BA)
print(f"Is AB != BA? {is_not_equal}")
AB_T = np.dot(A, B).T
BT_AT = np.dot(B.T, A.T)
print("\n3. Prove (AB)^T = B^T * A^T")
print("(A * B)^T:\n", AB_T)
print("B^T * A^T:\n", BT_AT)
is_equal_transpose = np.allclose(AB_T, BT_AT)
print(f"Is (AB)^T numerically close to B^T * A^T? {is_equal_transpose}")
A_lin = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3, 1, -1]])
B_lin = np.array([-1, -3, 9]) # B is the column vector of constants

print("\nCoefficient Matrix A:\n", A_lin)
print("Constant Vector B:\n", B_lin)
A_lin_inv = np.linalg.inv(A_lin)
X_inverse = np.dot(A_lin_inv, B_lin)
print("\nSolution using Inverse Method (X = A^-1 * B):")
print("Inverse of A (A^-1):\n", A_lin_inv)
print("Solution Vector X (x, y, z):\n", X_inverse)
print(f"The solution is: x = {X_inverse[0]:.0f}, y = {X_inverse[1]:.0f}, z = {X_inverse[2]:.0f}")
X_solve = np.linalg.solve(A_lin, B_lin)
print("\nSolution using np.linalg.solve(A, B):")
print("Solution Vector X (x, y, z):\n", X_solve)
print(f"The solution is: x = {X_solve[0]:.0f}, y = {X_solve[1]:.0f}, z = {X_solve[2]:.0f}")
B_check = np.dot(A_lin, X_solve)
print("\nVerification (A * X_solve should equal B):\n", B_check)
print(f"Does A * X_solve numerically equal B? {np.allclose(B_check, B_lin)}")
```
