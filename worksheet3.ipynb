```python
#3.2
#Question 1
df['fare_per_year'] = df['Fare'] / df['Age']
df = df[df['Age'] > 0]
high_fare_age = df[df['fare_per_year'] > 5]
high_fare_age_srt = high_fare_age.sort_values(
    by='fare_per_year',
    ascending=False
)
result = high_fare_age_srt[['Name', 'fare_per_year']]
result.head()

```

```python
df['fare_per_class'] = df['Fare'] / df['Pclass']
adult_males = df[(df['Sex'] == 'male') & (df['Age'] >= 18)]
adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)
result = adult_males_srt[['Name', 'Age', 'fare_per_class']]
result.head()
```

```python
# Question 1

# Total fare
total_fare = df['Fare'].sum()

# Fare by passenger class
fare_class_1 = df[df['Pclass'] == 1]['Fare'].sum()
fare_class_2 = df[df['Pclass'] == 2]['Fare'].sum()
fare_class_3 = df[df['Pclass'] == 3]['Fare'].sum()

# Fare distribution in percentage
fare_distribution = {
    'First Class (%)': (fare_class_1 / total_fare) * 100,
    'Second Class (%)': (fare_class_2 / total_fare) * 100,
    'Third Class (%)': (fare_class_3 / total_fare) * 100
}

fare_distribution
```

```python
# 3.3
# Question 2

def age_group(age):
    if age < 18:
        return 'Child'
    elif age < 65:
        return 'Adult'
    else:
        return 'Senior'
df = df.dropna(subset=['Age'])
df['age_group'] = df['Age'].apply(age_group)
total_passengers = len(df)
age_group_counts = df['age_group'].value_counts()
age_group_percentage = (age_group_counts / total_passengers) * 100

age_group_percentage

```
